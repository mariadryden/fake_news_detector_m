```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/hannadinizcunha/Downloads/WELFake_Dataset-3.csv', quotechar='"')
```

```python
df = df[['title', 'text', 'label']]
```

```python
df
```

```python
df = df.drop_duplicates(ignore_index=True)
```

```python
df = df.dropna()
```

```python
df
```

```python
total = []

for n in df['text']:
    total.append(len(n.split(" ")))
    
sum(total)/len(df['text'])

df['total'] = total
```

```python
df.sort_values('total').head(50)
```

```python
df = df[df['total'] > 100]
```

```python
df = df.reset_index(drop=True)
```
