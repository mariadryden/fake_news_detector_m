```python
import re
import string
import numpy as np
import pandas as pd
import seaborn as sns
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords, names
from nltk.stem import WordNetLemmatizer
```

```python
true= pd.read_csv('data/True.csv')
fake= pd.read_csv('data/Fake.csv')
```

```python
#Cleaning Data

true.drop_duplicates(inplace=True)
fake.drop_duplicates(inplace=True)

t_count=[]
for n in true['text']:
    t_count.append(len(n.split()))

true['count']= t_count
true= true[true['count'] > 500]


f_count=[]
for n in fake['text']:
    f_count.append(len(n.split()))

fake['count']= f_count
fake= fake[fake['count'] > 500]

  #Dropping columns

true= true.drop(columns=['title', 'subject', 'date', 'count'])
fake= fake.drop(columns=['title', 'subject', 'date', 'count'])

#Adding 'isfake'

true['isfake']= 0
fake['isfake']= 1

#Combining df

data= pd.concat([true, fake], axis=0, ignore_index=True)
data= data.reset_index()
```

```python
def cleaning(query):

    pattern = r'(?:[A-Z]+(?:[ \/][A-Z]+)* )?\(Reuters\) - '
    query= re.sub(pattern, '', query)

    query = query.strip()

    query = ''.join(char for char in query if not char.isdigit())

    for punctuation in string.punctuation:
        query = query.replace(punctuation, '')


    tokenized_query = word_tokenize(query)
    stop_words = set(stopwords.words('english'))

    tokenized_query_cleaned = [
        w for w in tokenized_query if not w in stop_words
    ]

    m_name= names.words('male.txt')
    tokenized2_query_cleaned = [
        w for w in tokenized_query_cleaned if not w in m_name
    ]

    f_name= names.words('female.txt')
    tokenized3_query_cleaned = [
        w for w in tokenized2_query_cleaned if not w in f_name
    ]

    lemmatized = [
        WordNetLemmatizer().lemmatize(word, pos = "v")
        for word in tokenized3_query_cleaned
    ]

    cleaned_query= ' '.join(word for word in lemmatized)

    cleaned_query = cleaned_query.lower()

    return cleaned_query
```

```python
data['text']= data['text'].apply(cleaning)
```

```python
def toothpick(text):
    twit = r'pictwittercom[a-z]+ '
    text= re.sub(twit, '', text)

    getty = r'[a-z]+getty image'
    text= re.sub(getty, '', text)

    com = r'[a-z]+com'
    text= re.sub(com, '', text)

    pattern = ['us ',
               'applo ',
               ' foxandfriends',
               'un ',
               'fraga',
               'twitter ',
               'facebook ',
               ' via',
               '’  ', ' " ', '" ',
               ]

    for p in pattern:
        text= re.sub(p, '', text)

    return text
```

```python
def names(text):

    name = ['kuczynski ',
               'donald ',
               'trump ',
               'letterman',
               'john ',
               'mccain ',
               'peter ',
               'mike ',
               'pence ',
               'palin ',
               'fivethirtyeight ',
               'obama ',
               'i '

               ]

    for n in name:
        text= re.sub(n, '', text)

    return text
```
